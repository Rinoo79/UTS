```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('sabun.jpg')
```

```python
cv2_imshow(image)
```

```python
result = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(result)
```

```python
ret, thresh1 = cv2.threshold(result, 50, 34, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(result, 50, 34, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(result, 50, 34, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(result, 50, 34, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(result, 50, 34, cv2.THRESH_TOZERO_INV)
```

```python
cv2_imshow(thresh1)
cv2_imshow(thresh2)
cv2_imshow(thresh3)
cv2_imshow(thresh4)
cv2_imshow(thresh5)
```
